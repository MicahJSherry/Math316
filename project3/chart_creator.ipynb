```python
# Generate all charts once and save them as JSON files
import json
import os

# Create a charts directory if it doesn't exist
os.makedirs('charts', exist_ok=True)

# Generate and save confusion matrix
confusion_chart = weather_classification_model(visual_type="confusion_matrix")
with open('charts/confusion_matrix.json', 'w') as f:
    json.dump(confusion_chart.to_dict(), f)

# Generate and save feature importance
feature_chart = weather_classification_model(visual_type="feature_importance")
with open('charts/feature_importance.json', 'w') as f:
    json.dump(feature_chart.to_dict(), f)

# Generate and save geographic performance
geo_chart = weather_classification_model(visual_type="geographic_performance")
with open('charts/geographic_performance.json', 'w') as f:
    json.dump(geo_chart.to_dict(), f)

print("All charts saved successfully!")
```
